Equation for a falling sphere with a mass of 10kg and a radius of 1m

The air drag on any object is given by $V'=\alpha v^2+\beta$ where $\beta$ constaines Gravity and $\alpha$ contains drag.$\newline$
To caculate alpha we will use the formula
$$
\frac{1}{2m} C_d \rho A\newline
C_d\approx0.47\newline
\rho\approx1.2250\newline
A=\pi r^2
$$
Since we are on earth $\beta = 9.817$ but on a diffrent planet is it just the force of gravity.$\newline$
In this case $\alpha \approx 0.0904$ and $\beta = 9.817$

Defining Variables in  the code

$radius(r) = 1\newline$
$mass(m) = 10\newline$
$C_d(Cd) = 0.47\newline$
$air pressure(\rho)=1.2250\newline$
$area(A) = \pi r^2\newline$
$\newline\newline \alpha = \frac{1}{2 m} C_d \rho A$

```python
#importing libraires
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
r=1
m=10
Cd=0.47
p=1.2250
A=np.pi*r**2

a=1/(2*m)*Cd*p*A
```

```python
#a is air drag and gravity is in b
b=-9.817
def dvdt(v,t):
    return a*v**2+b
v0=0
t=np.linspace(0,5,1000)
```

Using odeint to solve for dvdt and plotting the solutition

```python
solution = odeint(dvdt, v0, t)
```

```python
plt.plot(t,solution,label="Velocity")
plt.legend()
plt.grid()
```
